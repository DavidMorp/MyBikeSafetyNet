```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
```

```python
from azureml import Workspace

ws = Workspace()
ds = ws.datasets['bike_theft_data_updated.csv']
frame = ds.to_dataframe()
df = frame[['Latitude', 'Longitude']]
```

```python
plt.scatter(df['Longitude'],df['Latitude'], s=50)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10)
kmeans.fit(df)
y_kmeans = kmeans.predict(df)
```

```python
plt.scatter(df['Longitude'],df['Latitude'], c=y_kmeans, s=50, cmap='viridis')
plt.title("k-means algorithm on bike theft data")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
```
